```python
# Import required libraries
import matplotlib.pyplot as plt

# Sample data (simulating website traffic by source)
sources = ['Direct', 'Search', 'Social', 'Referral', 'Ads']
visits = [1200, 3500, 2000, 800, 1500]

# Create the bar chart
plt.figure(figsize=(8,5))
bars = plt.bar(sources, visits, color='skyblue')

# Add title and labels
plt.title("Website Traffic by Source")
plt.xlabel("Traffic Source")
plt.ylabel("Number of Visits")

# Add value labels above bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 50, str(height), ha='center', va='bottom')

# Display the chart
plt.show()
```

```python
# Import required libraries
import matplotlib.pyplot as plt

# Sample data (Average Monthly Rainfall in mm)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rainfall = [78, 60, 85, 120, 150, 200, 220, 210, 180, 130, 90, 70]

# Define seasons and colors
# Winter: Dec-Feb (blue), Spring: Mar-May (green), Summer: Jun-Aug (orange), Autumn: Sep-Nov (brown)
colors = ['blue', 'blue', 'green', 'green', 'green', 'orange',
          'orange', 'orange', 'brown', 'brown', 'brown', 'blue']

# Create the bar chart
plt.figure(figsize=(10,6))
bars = plt.bar(months, rainfall, color=colors)

# Add title and labels
plt.title("Average Monthly Rainfall")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")

# Add value labels above bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 5, str(height), ha='center', va='bottom')

# Display the chart
plt.show()
```
