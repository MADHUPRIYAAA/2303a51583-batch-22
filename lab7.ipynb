```python
import matplotlib.pyplot as plt

# Sample data for CO2 emissions over time
years = [1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020, 2024]
co2_emissions = [945, 1023, 1102, 1180, 1295, 1372, 1448, 1510, 1623, 1756, 1875, 1992, 2105, 2178]

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(years, co2_emissions, marker='o', color='blue', linestyle='-', linewidth=2)

# Add labels and title
plt.title("Global CO₂ Emissions Over Time", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("CO₂ Emissions (Million Metric Tons)", fontsize=12)

# Add grid and customize axis
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Show the chart
plt.show()

# Simple analysis of growth/reduction trend
if co2_emissions[-1] > co2_emissions[0]:
    print("✅ CO₂ emissions have increased over the years, indicating a growth trend.")
else:
    print("✅ CO₂ emissions have decreased over the years, indicating a reduction trend.")
```

```python
# Import required libraries
import matplotlib.pyplot as plt

# Sample data (simulating student performance over exams)
exam_numbers = [1, 2, 3, 4, 5]  # Exam numbers
student_scores = [65, 70, 75, 80, 85]  # Corresponding scores

# Create the line chart
plt.plot(exam_numbers, student_scores, marker='o', linestyle='-', color='b', label='Student Score')

# Add title and labels
plt.title("Student Performance Over Exams")
plt.xlabel("Exam Number")
plt.ylabel("Score")
plt.xticks(exam_numbers)  # Ensure x-axis shows each exam number
plt.yticks(range(0, 101, 10))  # Score range from 0 to 100
plt.grid(True)
plt.legend()

# Display the chart
plt.show()
```
