```python
import pandas as pd

# Step 1: Create sample datasets
customers_data = {'customer_id': [1, 2, 3, 4, 5],
                  'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
                  'city': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Los Angeles']}
customers_df = pd.DataFrame(customers_data)

transactions_data = {'transaction_id': [101, 102, 103, 104, 105, 106, 107, 108],
                     'customer_id': [1, 2, 1, 3, 4, 5, 2, 3],
                     'amount': [50.00, 75.00, 30.00, 100.00, 25.00, 150.00, 40.00, 60.00]}
transactions_df = pd.DataFrame(transactions_data)

print("Customers DataFrame:")
display(customers_df)
print("\nTransactions DataFrame:")
display(transactions_df)
```
